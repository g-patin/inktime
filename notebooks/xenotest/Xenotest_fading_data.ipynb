# Class XT 

```python
sample_nb = 'CAL01'

df = Class_XT.XT(sample_nb)
df = df.get_data(type='dE',patch='all')
df
```

```python
patches = ['M01','M02','M03','M04','M05','M06','M07','M08','M09','M10','M11']

fig, ax = plt.subplots(1,1,figsize=(15,10))
sns.set()

for patch in patches:
    data = df[df['patch','value']==patch]
    
    t = data['cum_hrs']['value']
    dE = data['dE00']['mean']

    ax.plot(t,dE, label=patch)
    
    
ax.set_xlabel('Cumulative MJ/m$^2$')
ax.set_ylabel('$\Delta E^*_{00}$')
    
plt.legend()
plt.tight_layout()
plt.show()
```
